```python
sd = pd.read_csv("/content/gdrive/MyDrive/Smoke_Detection/smoke_detection_iot.csv")
sd
```

```python
sd.drop('Index',axis=1,inplace =True)
```

```python
sd['Temperature'] = sd["Temperature[C]"]
sd['Humidity'] = sd['Humidity[%]'] 
sd['TVOC'] = sd['TVOC[ppb]'] 
sd['Pressure'] = sd['Pressure[hPa]'] 
sd['eCO2'] = sd['eCO2[ppm]']
sd
```

```python
sd.drop(['Temperature[C]','Humidity[%]','TVOC[ppb]','eCO2[ppm]','Pressure[hPa]'],axis=1,inplace=True)
sd
```

```python
sd.drop('TVOC[ppb',axis=1,inplace=True)
```

```python
sd
```

```python
x = pd.DataFrame(sd.drop('Fire Alarm',axis=1))
y = pd.DataFrame(sd['Fire Alarm'])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
#test_size = 0.2 means 20:80 ratio of splitting test and training dataset
```
